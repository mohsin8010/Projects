```python
import requests
from pandas.io.json import json_normalize
URL = "https://api.covid19india.org/data.json"
data = requests.get(url=URL).json()
covid19_df = json_normalize(data['statewise'])
print("Total Confirmed Cases: "+str(covid19_df[covid19_df.state == "Total"]['confirmed']))
print("Total Active Cases: "+str(covid19_df[covid19_df.state == "Total"]['active']))
print("Total Recovered Cases: "+str(covid19_df[covid19_df.state == "Total"]['recovered']))
print("Total Deceased Cases: "+str(covid19_df[covid19_df.state == "Total"]['deaths']))
print(covid19_df[['state','confirmed','active','recovered','deaths']].sort_values(by="confirmed", ascending=False))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
fig_size=plt.rcParams["figure.figsize"]
fig_size[0]=15
fig_size[1]=15
plt.rcParams["figure.figsize"]=fig_size
```

```python
df=covid19_df[['state','confirmed','active','recovered','deaths']].sort_values(by="confirmed")
df.sort_values(["confirmed"],ascending=True)
state_wise_cases=df["confirmed"]

state_name=df["state"]
plt.barh(state_name,state_wise_cases)
plt.title("State-Wise-Confirmed-Cases")
plt.ylabel("States")
plt.xlabel("Number of Cases")
plt.show()
```

```python
data=df
data = data.drop([0], axis=0)
no_of_cases=data["confirmed"]
active_cases=data["active"]
recovered_cases=data["recovered"]
no_of_death=data["deaths"]
plt.barh(active_cases,no_of_cases,label="Active",color='b')
plt.barh(recovered_cases,no_of_cases,label="Recovered",color='g')
plt.barh(no_of_death,no_of_cases,label="Dead",color='r')
plt.legend()
plt.title("Corona Cases")
plt.xlabel("Confirmed Cases")
plt.ylabel("Recovered,Active,Deceased Cases")
plt.show()
```
